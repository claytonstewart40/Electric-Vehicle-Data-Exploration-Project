```python
import pandas as pd
csv_path = "C:/Users/djgoo/DataAnalysis/Project-3/EV Population Data/Electric_Vehicle_Population_Data.csv"
mapsdf = pd.read_csv(csv_path)
mapsdf
```

```python
mapsdf["Vehicle Location"] = mapsdf["Vehicle Location"].str.replace('POINT ', "")
```

```python
mapsdf["Vehicle Location"] = mapsdf["Vehicle Location"].str.replace(' ', ", ")
```

```python
mapsdf["Vehicle Location"] = mapsdf["Vehicle Location"].str.replace('(', "")
```

```python
mapsdf["Vehicle Location"] = mapsdf["Vehicle Location"].str.replace(')', "")
```

```python
columns = ["VIN (1-10)","County","City", "Postal Code", "Vehicle Location", "Make", "Model"]
mapsdf_cleaned = mapsdf[columns].copy()
mapsdf_cleaned
```

```python
mapsdf_cleaned[['Latitude', 'Longitude']] = mapsdf_cleaned["Vehicle Location"].str.split(',', expand=True)
```

```python
mapsdf_cleaned.drop("Vehicle Location", axis=1, inplace=True)
```

```python
mapsdf_cleaned.to_json("mapsdatacleaned.json", orient="split", index=False)
```
